 Теперь необходимо понять, что такое изображение в компьютерном представлении. Изображение - это не что иное, как стандартный массив **Numpy**, содержащий пиксели точек данных. Чем больше пикселей в изображении, тем лучше его разрешение. Вы можете считать пиксели крошечными блоками информации, расположенными в виде двумерной сетки, а глубина пикселя относится к информации о цвете, присутствующей в нем. Чтобы быть обработанным компьютером, изображение должно быть преобразовано в двоичную форму. Цвет изображения можно рассчитать следующим образом:
 
      Number of colors/ shades = 2^bpp где bpp представляет биты на пиксель.
   
  Например, если у нас 1 бита/пиксель, то возможно представление 2х цветов, 4 бита/пиксель - 16 возможных цветов.
  
  **Какие могут быть изображения**:
  
  - Бинарные. Тут все просто. У нас возможно только два варианта: 0 или 1 (черный или белый)
      
      
  - Изображения в серых оттенках. Изображение в градациях серого состоит из 8 бит на пиксель. Это означает, что он может иметь 256 различных оттенков, где 0 пикселей будут представлять черный цвет, а 255 - белый. Например, изображение ниже показывает изображение в градациях серого, представленное в виде массива. Изображение в градациях серого имеет только 1 канал, где канал представляет размерность.
  
     
     
  - Цветные изображения  - представляют собой комбинацию тех цветов: синего, красного и зеленого.
  
  
  Цветные изображения также состоят из 8 бит на пиксель. Однако тут мы имеем три пространства

Как видим, изображение было трансформировано в массив размером 220х220 и имеет 3 канала.
Выведем изображение:


```python
plt.imshow(img_raw)
```

картинка отображается не так как мы ее видели, так как OpenCV читает изображение в формате BGR, а matplotlib - RGB.
Подправим немного

```python
img_rgb = cv2.cvtColor(img_raw, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)#[:100,:100,0]
```

https://opencv-python-tutroals.readthedocs.io/en/latest/py_tutorials/py_imgproc/py_thresholding/py_thresholding.html
использование трешхолда:

```python
ret,thresh3 = cv2.threshold(img_rgb,127,255,cv2.THRESH_TRUNC)
plt.imshow(thresh3)
```

Для распознания лица необходимо привести картинку к серым тонам:


```python
test_image_gray = cv2.cvtColor(img_raw, cv2.COLOR_BGR2GRAY)
lena_rgb = cv2.cvtColor(test_image_gray, cv2.COLOR_BGR2RGB)
plt.imshow(lena_rgb, cmap='gray')
```

Обнаружение лица осуществляется с помощью **классификаторов**. **Классификатор** - это по сути алгоритм, который решает, является ли данное изображение положительным (лицо) или отрицательным (не лицо). Классификатор должен быть обучен на тысячах изображений с лицами и без лиц. 

OpenCV поставляется с множеством предварительно обученных классификаторов. Например, существуют классификаторы улыбки, глаз, лица и т.д. Они представлены в виде XML-файлов и находятся в [папке](https://github.com/opencv/opencv/tree/master/data/haarcascades). Однако, чтобы упростить задачу, можно загрузить XML-файлы и поместите их в папку данных в том же рабочем каталоге, что и код.

[Скачать файл](https://raw.githubusercontent.com/opencv/opencv/master/data/haarcascades/haarcascade_frontalface_default.xml)

В данном примере исспользуем классификатор Каскада Хаара (Haar Classifier). Как он работает можно посмотреть тут http://www.willberger.org/cascade-haar-explained/    
https://habr.com/ru/company/recognitor/blog/228195/    


В **OpenCV** есть класс **CascadeClassifier**.
Исспользуем  метод **detectMultiScale** классификатора. Этот метод вернет прямоугольник с координатами (x, y, w, h) вокруг обнаруженного лица. **detectMultiScale** имеет два  параметра, которые должны быть настроены в соответствии с данными:

- scalefactor На групповой фотографии могут быть лица,находящиеся ближе к камерой, чем другие. Естественно, такие лица выглядят более заметными, чем те, что стоят позади. Этот фактор компенсирует это.

- minNeighbors Этот параметр указывает   на количество соседей, которые прямоугольник должен называть гранью. 

Вы можете прочитать больше об этом методе [здесь](https://docs.opencv.org/2.4.13.2/modules/objdetect/doc/cascade_classification.html#cv2.CascadeClassifier.detectMultiScale)



```python
haar_cascade_face = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
faces_rects = haar_cascade_face.detectMultiScale(test_image_gray, scaleFactor = 1.2, minNeighbors = 5)
print('Faces found: ', len(faces_rects))
faces_rects
```

похоже лицо найдено, и хорошо бы как-то увидеть результат. Нарисуем прямогульник на основании полученных координат.
используя OpenCV можно рисовать линии, круги и разумеется прямоугольники). Как обычно, для прорисовки необходимо задать параметры метода rectangle: координаты верхней левой и нижней правой точки, дальше задается цвет линии и толщина


```python
for (x,y,w,h) in faces_rects:
     cv2.rectangle(lena_rgb, (x, y), (x+w, y+h), (0, 255, 0), 2)

plt.imshow(lena_rgb)
```
